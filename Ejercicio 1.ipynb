```python
def agregar_cliente(base_datos):
    rfc = input("Ingrese el RFC del cliente: ")
    nombre = input("Ingrese el nombre del cliente: ")
    direccion = input("Ingrese la dirección del cliente: ")
    telefono = input("Ingrese el teléfono del cliente: ")
    correo = input("Ingrese el correo del cliente: ")
    preferente = input("¿Es un cliente preferente? (SI/NO): ")
    
    if preferente == "SI":
        preferente = True
    elif preferente == "NO":
        preferente = False
    else:
        print("Opción inválida. Por favor, seleccione SI en caso de ser preferente o NO en caso de no serlo.")
    
    cliente = {
        "nombre": nombre,
        "dirección": direccion,
        "teléfono": telefono,
        "correo": correo,
        "preferente": preferente
    }
    
    base_datos[rfc] = cliente
    print("Cliente agregado correctamente.")

def eliminar_cliente(base_datos):
    rfc = input("Ingrese el RFC del cliente que desea eliminar: ")
    if rfc in base_datos:
        del base_datos[rfc]
        print("Cliente eliminado correctamente.")
    else:
        print("No se encontró ningún cliente con ese RFC.")

def mostrar_cliente(base_datos):
    rfc = input("Ingrese el RFC del cliente que desea mostrar: ")
    if rfc in base_datos:
        cliente = base_datos[rfc]
        print("RFC:", rfc)
        for key, value in cliente.items():
            print(key.capitalize() + ":", value)
    else:
        print("No se encontró ningún cliente con ese RFC.")

def listar_clientes(base_datos):
    print("Lista de todos los clientes:")
    for rfc, cliente in base_datos.items():
        print("RFC:", rfc)
        print("Nombre:", cliente["nombre"])
        print("--------")

def listar_clientes_preferentes(base_datos):
    print("Lista de clientes preferentes:")
    for rfc, cliente in base_datos.items():
        if cliente["preferente"]:
            print("RFC:", rfc)
            print("Nombre:", cliente["nombre"])
            print("--------")

def main():
    base_datos = {}
    opcion = 0
    
    while opcion != 6:
        print("Menú:")
        print("(1) Añadir cliente")
        print("(2) Eliminar cliente")
        print("(3) Mostrar cliente")
        print("(4) Listar todos los clientes")
        print("(5) Listar clientes preferentes")
        print("(6) Terminar")
        
        opcion = int(input("Seleccione una opción: "))
        
        if opcion == 1:
            agregar_cliente(base_datos)
        elif opcion == 2:
            eliminar_cliente(base_datos)
        elif opcion == 3:
            mostrar_cliente(base_datos)
        elif opcion == 4:
            listar_clientes(base_datos)
        elif opcion == 5:
            listar_clientes_preferentes(base_datos)
        elif opcion == 6:
            print("Programa terminado.")
        else:
            print("Opción inválida. Por favor, seleccione una opción del menú.")

main()
```
